```python
%reload_ext autoreload
%autoreload 2

import numpy as np

%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import colormaps as cms
mpl.rc_file(f"../matplotlibrc")
```

```python
# phot
l = np.array([30., 30., 29., 28., 28., 27., 27., 27., 27., 26., 26., 26.])
multiplier = 1/3
x = np.arange(len(l))
mass = np.logspace(2, 12, 11) # from script

p = np.polyfit(x, l, 1)
lsmooth = np.polyval(p, x)
lsmooth_adjusted = lsmooth + np.log10(multiplier)

x = x[:-1]
lsmooth = lsmooth[:-1]
lsmooth_adjusted = lsmooth_adjusted[:-1]
```

```python
# more phot
xmore = np.arange(0.5, len(l)-0.5-0.001, 1)
massmore = 10**(xmore + 2)
lsmoothmore = np.polyval(p, xmore)
lsmoothmore_adjusted = lsmoothmore + np.log10(multiplier)

xmore = xmore[:-1]
massmore = massmore[:-1]
lsmoothmore = lsmoothmore[:-1]
lsmoothmore_adjusted = lsmoothmore_adjusted[:-1]
```

```python
# elec
l = np.array([27., 26., 27., 28., 28., 27., 27., 26., 26., 26., 25., 25.]) + 1.
multiplier = 4
x = np.arange(len(l))

p = np.polyfit(x, l, 3)
lsmooth = np.polyval(p, x)
lsmooth_adjusted = lsmooth + np.log10(multiplier)
```

```python
plt.plot(mass, lsmooth_adjusted, "s", color='C0')
plt.plot(massmore, lsmoothmore_adjusted, "s", color='C1')
plt.xscale('log')
plt.axvline(1e2, color='k', ls='--')
plt.axvline(1e12, color='k', ls='--')
```

```python
plt.plot(x, l)
plt.plot(x, lsmooth, "-+", lw=1)
plt.plot(x, lsmooth_adjusted, "-+", lw=1)
plt.plot(xmore, lsmoothmore, "s", color='C1')
plt.plot(xmore, lsmoothmore_adjusted, "s", color='C2')
```
