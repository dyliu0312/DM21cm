```python
DIR = '/zfs/yitians/ionhist/ionhist/maketf/tfdata/array/nhd_test'
```

```python
hep_tf = np.load(DIR+'/hep_tf.npy')
```

```python
hep_tf.shape
```

```python
abscs_nhd = {
    'nhd' : np.linspace(-1., 1.7, 5),
    'x' : np.linspace(1e-5, 1-1e-5, 5),
    'rs' : np.logspace(np.log10(5.), np.log10(50.), 5),
    'eng' : abscs['photeng']
}
```

```python
abscs_nhd['x']
```

```python
x_i = 2
rs_i = 4

fig, axs = plt.subplots(3, 5, figsize=(25, 20))

extent = np.log10((abscs['photeng'][0],abscs['photeng'][-1], abscs['photeng'][0], abscs['photeng'][-1]))

for nhd_i in range(5):
    axs[0][nhd_i].imshow(plot_val(hep_tf[nhd_i,x_i,rs_i]), vmin=-55, vmax=10, extent=extent, cmap='plasma')
    axs[0][nhd_i].set(xlabel='$\log_{10}E_\mathrm{out}$', ylabel='$\log_{10}E_\mathrm{in}$')
    axs[0][nhd_i].text(0, -1, f"x={abscs_nhd['x'][x_i]:.3f}\nrs={abscs_nhd['rs'][rs_i]:.3f}", color='white')
    axs[0][nhd_i].set(title=f"nhd={abscs_nhd['nhd'][nhd_i]:.3f}")
    
nhd_i = 1
rs_i = 4

for x_i in range(5):
    axs[1][x_i].imshow(plot_val(hep_tf[nhd_i,x_i,rs_i]), vmin=-55, vmax=10, extent=extent, cmap='plasma')
    axs[1][x_i].set(xlabel='$\log_{10}E_\mathrm{out}$', ylabel='$\log_{10}E_\mathrm{in}$')
    axs[1][x_i].text(0, -1, f"nhd={abscs_nhd['nhd'][nhd_i]:.3f}\nrs={abscs_nhd['rs'][rs_i]:.3f}", color='white')
    axs[1][x_i].set(title=f"x={abscs_nhd['x'][x_i]:.3f}")
    
nhd_i = 1
x_i = 2

for rs_i in range(5):
    axs[2][rs_i].imshow(plot_val(hep_tf[nhd_i,x_i,rs_i]), vmin=-55, vmax=10, extent=extent, cmap='plasma')
    axs[2][rs_i].set(xlabel='$\log_{10}E_\mathrm{out}$', ylabel='$\log_{10}E_\mathrm{in}$')
    axs[2][rs_i].text(0, -1, f"x={abscs_nhd['x'][x_i]:.3f}\nnhd={abscs_nhd['nhd'][nhd_i]:.3f}", color='white')
    axs[2][rs_i].set(title=f"rs={abscs_nhd['rs'][rs_i]:.3f}")
```
