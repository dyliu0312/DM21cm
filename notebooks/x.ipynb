```python
import numpy as np
import os



%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc_file('../matplotlibrc')
```

```python
from PIL import Image, ImageOps # just for fun, you don't need this

def get_random_icon(size=None, prefix='../../resources/icons'):
    fn = np.random.choice([f for f in os.listdir(prefix) if f.endswith('.png')])
    im = Image.open(f'{prefix}/{fn}')
    if size is not None:
        im = im.resize((size, size))
    im = np.array(im, dtype=np.float32)
    im = np.dot(im, [0.299, 0.587, 0.114, 0.]) * im[:,:,-1]
    return im / np.max(im)
```

```python
plt.imshow(get_random_icon(size=30))
```
